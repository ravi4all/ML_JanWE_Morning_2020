```python
# Data Analysis
import pandas as pd
# Data Visualization
import matplotlib.pyplot as plt
# Styling of Graphs
import seaborn as sns
plt.style.use('tableau-colorblind10')
```

```python
matches = pd.read_csv('ipldata_19/matches.csv')
deliveries = pd.read_csv('ipldata_19/deliveries.csv')
```

```python
pd.value_counts(matches['season'])
```

```python
plt.figure(figsize=(10,6))
# plt.hist(matches['season'])
plt.style.use('grayscale')
sns.countplot(x='season', data=matches)
plt.title("Matches Per Season")
plt.xlabel("IPL Seasons")
plt.ylabel("Number of Matches")
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='winner', data=matches)
plt.title("Total Wins")
plt.xlabel("Team Name")
plt.ylabel("Number of Wins")
plt.xticks(rotation='vertical')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='season',data=matches,hue='toss_decision')
plt.xticks(rotation='vertical')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.style.use('ggplot')
ax = matches['player_of_match'].value_counts().head(10).plot.bar(width=0.9,color=sns.color_palette('inferno'))    
for patch in ax.patches:
#     print(patch)
    ax.annotate(format(patch.get_height()),(patch.get_x() + 0.3, patch.get_height() + 0.3))
```

```python
plt.figure(figsize=(10,6))
ax = deliveries.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10).plot.bar(width=0.9,color=sns.color_palette('inferno'))    
for patch in ax.patches:
    ax.annotate(format(patch.get_height()),(patch.get_x() + 0.3, patch.get_height() + 10))
```
